```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/moosaabbasi/ml/insurance valuable.csv')
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('charges',axis=1)
y = df['charges']

X_train,X_test,Y_train,Y_test = train_test_split(x ,y ,test_size = 0.20,random_state = 42)
```

```python


model = LinearRegression(fit_intercept=False, 
    n_jobs=-1,
    copy_X=True)
model.fit(X_train , Y_train)


```

```python
prediction = model.predict(X_test)
print(f"prediction after the training\n {prediction[:5]}")
print(f"actual charges \n {Y_test[:5]}")
```

```python
#Geanrating a plot for checking the preiction
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6))
plt.scatter(prediction , Y_test , color = 'red' , alpha = 0.7)
plt.plot([Y_test.min(),Y_test.max() ],[Y_test.min(),Y_test.max()],color="black")
plt.xlabel("Actual value")
plt.ylabel("preicted value")
plt.show()
```
